```python
import numpy as np
from scipy.stats import t
```

```python
vector = [0.588,	0.589,	0.608,	0.608,	0.607,	0.608,	0.594,	0.59	,0.586,	0.58]

alpha = 0.05
```

```python
def doubleCI(vector, alpha):                    # function for Confidence Intervals computation
    mean = np.mean(vector)                      # compute mean
    stde = np.std(vector, ddof=1)               # compute standard deviation
    n = len(vector)                             # get number of elements of vector
    talpham = t.ppf(1-alpha/2, df=n-1)          # compute t-test with alpha/2
    lower = mean - talpham * stde / np.sqrt(n)  # compute lower bound of CI
    upper = mean + talpham * stde / np.sqrt(n)  # compute upper bound of CI
    return lower, mean, upper                   # return lower bound, mean and upper bound
```

```python
lower, mean, upper = doubleCI(vector, alpha)
```

```python
print([round(lower,4), round(mean,4), round(upper,4)])
```
